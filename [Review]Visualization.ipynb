```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
```

```python
#간단한 선그래프 그리기
x = np.arange(1, 5)
plt.plot(x)
plt.title('Line Plot')
plt.xlabel('XDATA')
plt.ylabel('YDATA')
plt.show()
x
```

```python
x = np.arange(1, 5)
y = [5, 20, 24, 25]
plt.plot(x,y, 'rx:')

plt.title('Line Plot2')
plt.xlabel('XDATA')
plt.ylabel('YDATA')
plt.show()
```

```python
data = np.arange(0, 5, 0.2)
plt.plot(data, data, 'r--', data, data**2, 'bs:', data, data**3, 'g^:')
plt.show()
```

```python
x = np.arange(10)
y = np.random.randint(0, 100, 10)
plt.plot(x, y, c='b', lw=3, ls=':', marker='v', ms=10, mec='r', mew=2, mfc='y')
plt.title('exp3')
plt.show()
```

```python
font = {
    'family' : 'serif',
    'color' : 'darkred',
    'weight' : 'normal',
    'size' : 16
}

x = np.linspace(0.0, 5.0, 100)
y = np.cos(2*np.pi*x)*np.exp(-x)

plt.plot(x, y, 'k')
plt.title('Damped exponential decal', fontdict = font)

plt.xlabel('time', fontdict=font)
plt.ylabel('voltage(mV)', fontdict=font)
plt.show()
```

```python
t = np.arange(0.01, 5.0, 0.01)
s = np.sin(2 * np.pi * t)

plt.plot(t,s)
plt.xlim(0.01, 5.0)
plt.ylim(-1.5, 1.5)
plt.margins(0.1)
plt.show()
```

```python
x = np.arange(1, 5)
y = [1, 4, 5, 9]
labels = ['f', 'j','a','s']
plt.plot(x, y)
plt.xticks(x, labels, rotation=45)
plt.margins(0.1) #눈금 마진 조절
plt.show()
```

```python
#범례

a = np.arange(0, 3, 0.02)
c = np.exp(a)
d = c[::-1]

fig, ax = plt.subplots()
ax.plot(a, c, 'k--', label='Model length')
ax.plot(a, d, 'k:', label='Data length')
ax.plot(a, c+d, 'k', label='Total message length')

legend = ax.legend(loc='upper center', shadow='True')

plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 256)
c, s = np.cos(x), np.sin(x)


plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1], ['Low', 'Zero', 'High'])

plt.plot(x, c, ls='--', label='cosine')
plt.plot(x, s, ls=':', label='sine')
plt.legend(loc=4)
plt.show()
```

```python
#figure

plt.figure(figsize=(10, 3))

x = [1, 3, 5, 8, 12]
y = np.power(x, 3)

plt.plot(x, y, 'go--')
plt.show()
```

```python
t = np.arange(0.0, 2.0, 0.01)
s1 = np.sin(2*np.pi*t)
s2 = np.cos(2*np.pi*t)

plt.subplot(2, 1, 1)
plt.title('Sin/Cos')
plt.plot(t, s1, 'y-')

plt.subplot(2, 1, 2)
plt.plot(t, s2, 'r-')
plt.xlabel('Times(s)')

plt.show()
```

```python
x = np.arange(0, 10, 0.01)

fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0,0].plot(x, np.sin(x))
axes[0,1].plot(x, np.cos(x))
axes[1,0].plot(x, np.tanh(x))
axes[1,1].plot(x, np.sin(x)**2)

plt.show()
```

```python
np.random.seed(0)

fig, ax0 = plt.subplots()
ax1 = ax0.twinx()

ax0.plot(np.random.rand(5), 'b-', label='y0')
ax0.set_title('TITLE')
ax0.set_xlabel('axis 2')
ax0.set_ylabel('first Y')

ax1.plot(np.random.rand(5), 'r:', label='y1')
ax1.set_ylabel('second Y')


plt.show()
```

```python
labels = ['Seoul', 'Busan', 'Incheon', 'Gwangju', 'Jeju']
y = [20, 34, 30, 35, 27]
x = np.arange(len(y))

plt.title('TITLe')
plt.bar(x, y)
plt.xticks(x, labels)

plt.xlabel('City')
plt.ylabel('Frequency')

plt.show()
```

```python
labels = ['Seoul', 'Busan', 'Incheon', 'Gwangju', 'Jeju']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels))
width= 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

ax.set_title('Scores by group and gender')
ax.set_xticks(x, labels) #plt.xticks(x, labels)
ax.legend()
plt.xlabel('City')
plt.ylabel('Scores') # ax.set_ylabel('Scores')


plt.show()
```

```python
labels = ['Seoul', 'Busan', 'Incheon', 'Gwangju', 'Jeju']
y = [20, 34, 30, 35, 27]
x = np.arange(len(y))
error = np.random.rand(len(labels))

plt.barh(x, y, xerr=error, color='#09acff', alpha=0.5)

plt.yticks(x, labels)
plt.ylabel('City')
plt.xlabel('Frequency')

plt.show()
```

```python
labels = ['Seoul', 'Daegu', 'Busan','Incheon']
sizes = [22, 15, 36, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0, 0.1, 0) #튀어나오게 함.

plt.pie(sizes, explode = explode, labels = labels, colors = colors, shadow=True, autopct='%1.1f%%', startangle = 90)
plt.axis('equal') #axis를 안줄경우 어떻게 보여지는지
plt.show()
```
