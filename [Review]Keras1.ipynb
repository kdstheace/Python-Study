```python
#1) package import
from keras.models import Sequential #데이터를 순차적으로 나오게
from keras.layers import Dense, Activation  #Activation은 활성화 함수
import tensorflow as tf
import numpy as np
```

```python
#2) 데이터 준비 
X = np.random.random((1000, 10))
y = np.random.randint(6, size=(1000, 1)) #sklearn 은 y인 정답은 1차원이었는데, keras에서는 x, y 둘다 이차원
```

```python
#4) 계층쌓기
model = Sequential()
model.add(Dense(units=32, activation='relu', input_dim=10))
#units: 그다음 출력 레이어 몇개?,    activation : 활성함수,     input_dim
model.add(Dense(units=1, activation='linear')) #두번째부터는input _dim안써

model.summary()  #dense2에서는 10개가 들어가서 32개 나오는데 bias 까지 생각하니 까 param#가 352개
```

```python
#5) 최적화
#loss함수, 최적화함수 선정
model.compile(optimizer='sgd', loss='mse', metrics=['accuracy']) #sgd: gradient descent, mse: mean squared error

#6) 모델학습
model.fit(X, y, epochs=15, batch_size=100, verbose = 1) 
#epochs : 학습을 몇번시킬 것인가? batch_size : 가져와서 학습하는 개수,  verbose=0, 1(공부하는 걸 지켜보겠다.)
```

```python
#Evaluate
score = model.evaluate(X, y, verbose=1)
print('오차 : ', score[0])
print('정확도 : ', score[1])
```

```python
#predict
pred = model.predict(X)
pred
```
