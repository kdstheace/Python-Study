# **1. Linear Regression Basic**
- Cost: 제곱 - 경사하강법법법
- y = wa+b (weight, bias-y절편) x는 feature
- 단변량 x 1개, 다변량 x여러개


```python
import tensorflow as tf
import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Cost Function 그리기
def cost(w, x, y):
  s = 0
  for i in range(len(x)):
    s += (w*x[i] - y[i])**2
  
  return s / len(x)


x = [1., 2., 3.] #feature
y = [2., 4., 6.] 

w_val = []
cost_val=[]

for i in range(-30, 71):
  w = 0.1 * i
  c = cost(w, x, y)
  w_val.append(w)
  cost_val.append(c)

print(w_val)
print(cost_val)

#--------cost값을 이용한 그래프 그리기
plt.plot(w_val, cost_val)
plt.xlabel('Weight')
plt.ylabel('Cost')
plt.show()
```

# **1.단변량 데이터의 Linear Regression**


```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

#1. 데이터 준비
X = np.array([[1], [3], [2], [4], [7], [4], [9], [2], [3], [2], [6], [3], [2], [7] ])
y = np.array([3, 9, 6, 7,10, 6, 12, 2, 4, 3, 8, 5, 3, 10 ])

#2. 모델 준비
model = LinearRegression()

#3. 학습(Fitting)
model.fit(X,y)

#4. 평가(evaluate)

#5. 시각화(visualize)
plt.plot(X,y,'o')
plt.show()

#6. 예측(predict)
X_tmp = [[7]]
y_tmp = model.predict(X_tmp)
print(y_tmp)
```

```python
#sklearn이용한 LinearRegression 코드의 사용  - 2
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

#1. 데이터 준비
X = np.random.rand(100, 1)*10
y = np.random.rand(100, 1)*5.4 + (X*2.3)

plt.plot(X,y, 'o')
plt.show()
#2. 모델 준비
model = LinearRegression()
#3. 학습
model.fit(X, y)
#4. 예측, 시각화
plt.plot(X, y, 'o')
plt.plot(X, model.predict(X))
plt.show()
```

```python
#문제: 키와 몸무게를 이용해 모델을 학습시키고 키 170인 사람의 몸무게를 예측하는 프로그램을 작성해라
raw_data = {'Height' : [165.67, 171.63, 169.5, 168.22, 183.5],
            'Weight' : [50.4, 58.45, 52.34, 50.5, 71.34 ],}

X = np.array(raw_data['Height']).reshape(-1,1) #행은 몰라행, 열은 1 무조건 이렇게 쓰는 버릇하자!
y = raw_data['Weight']

model = LinearRegression()
model.fit(X, y)
result = model.predict([[170]])
print(result)

plt.plot(X, y, 'o')
plt.plot(X, model.predict(X))
plt.show()
```
