# **1. Linear Regression Basic**
- Cost: 제곱 - 경사하강법법법
- y = wa+b (weight, bias-y절편) x는 feature

```python
import tensorflow as tf
import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Cost Function 그리기
def cost(w, x, y):
  s = 0
  for i in range(len(x)):
    s += (w*x[i] - y[i])**2
  
  return s / len(x)


x = [1., 2., 3.] #feature
y = [2., 4., 6.] 

w_val = []
cost_val=[]

for i in range(-30, 71):
  w = 0.1 * i
  c = cost(w, x, y)
  w_val.append(w)
  cost_val.append(c)

print(w_val)
print(cost_val)

#--------cost값을 이용한 그래프 그리기
plt.plot(w_val, cost_val)
plt.xlabel('Weight')
plt.ylabel('Cost')
plt.show()
```
