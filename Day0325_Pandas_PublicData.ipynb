# 공공데이터를 이용한 처리(서울시 공공데이터 자전거)

```python
from pandas import DataFrame, Series
import pandas as pd
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
path = 'drive/MyDrive/Colab Notebooks/data/'
data1 = '서울특별시 공공자전거 대여소 정보(19.12.9).xlsx'
data2 = '서울특별시 공공자전거 대여정보_201911_2.csv'

# mnist = pd.read_csv('sample_data/mnist_train_small.csv')
# mnist
```

```python
place = pd.read_excel(path+data1)
place
```

```python
# null이 들어가 있는지 확인
```

```python
#nan이 들어있는 컬럼 출력해보기
place[place['대여소ID'].isnull()]
```

```python
#전처리 작업하기: NaN값 삭제하기
place = place.dropna()
place
```

```python
#대여소 ID의 데이터를 float64 -> int32로 변경하기대여소 ID의 데이터를 float64 -> int32로 변경하기
place['대여소ID'].dtypes
```

```python
place = place.astype({'대여소ID' : np.int32})
```

```python
rent = pd.read_csv(path + data2, encoding='cp949', delimiter=',') #data2들어있는 정보중 한글이 utf-8로 되어 있으면 못읽음. encoding을 cp-949로
rent
```

```python
# rent데이터에 null이 있는지 확인 후 존재하면 drop시킬 것.
rent.isnull().sum()
```

```python
# 대여소 정보 중 중복된 데이터를 처리
place['대여소_구'].unique()
```

```python
# 전체 대여소 구 갯수
print('대여소를 운영중인 지자체 수: ', len(place['대여소_구'].unique()))
```

```python
# 대여소 구별 평균 거치대 수: mean
place.groupby('대여소_구')['거치대수'].mean()
```

```python
#agg()를 이용해 통계 내기
mydata = place.groupby('대여소_구')
mydata['거치대수'].agg([np.sum, np.mean, np.amax, np.amin])

#[문제]3번 줄에서 출력된 정보중 sum칼럼의 값들을 기준으로 역순 정리해서 출력하자.
mydata['거치대수'].agg([np.sum, np.mean, np.amax, np.amin]).sort_values(by='sum', ascending=False)
```

```python
#(대여소 정보merge대여정보) place테이블과 rent테이블 조인하기 id로
bicycle = pd.merge(place, rent, left_on='대여소ID', right_on='대여 대여소번호')
bicycle
```

```python
#[문제]agg함수를 이용해 각 구별 평균 자전거 이용시간, 최대 이용거리, 최소 이동거리
#bicycle 이용
a = bicycle.groupby('대여소_구')['이용시간'].agg([np.mean])
b = bicycle.groupby('대여소_구')['이용거리'].agg([np.amax, np.amin])
print(a)
print(b)
```
